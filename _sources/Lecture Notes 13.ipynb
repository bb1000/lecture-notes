```python
import pandas as pd
```

```python
df = pd.read_csv('contestants.csv')
```

```python
df
```

### How many contestants per year?

```python
year_count = df.groupby('year')['to_country'].count()
year_count
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(year_count.index, year_count)
```

```python
# !conda install -y seaborn
```

```python
import seaborn
```

```python
seaborn.relplot(data=year_count, kind='line')
```

## How many times each country participated per year

```python
country_year = df.groupby('to_country')['year'].count().reset_index().sort_values('year', ascending=False)
country_year
```

```python
seaborn.catplot(data=country_year, kind='bar', x='year', y='to_country', height=10)
```

## Winners?

```python
df.head()
```

```python
winners = df[df['place_final'] == 1]
```

```python
winners
```

```python
wins_country = winners.groupby('to_country')['place_final'].count().sort_values(ascending=False).reset_index()
wins_country
```

```python
seaborn.catplot(data=wins_country, kind='bar', x='place_final', y='to_country')
```
