### Dataframes

```python
data = {
    'country': ['Belgium', 'France', 'Germany', 'Netherlands', 'United Kingdom'],
    'population': [11.3, 64.3, 81.3, 16.9, 64.9],
    'area': [30510, 671308, 357050, 41526, 244820],
    'capital': ['Brussels', 'Paris', 'Berlin', 'Amsterdam', 'London']
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df['country']
```

```python
df.columns
```

```python
df.set_index('country')  # does not change original data
```

```python
df
```

to save changes

* reassign
* use `inplace` option

```python
# df = df.set_index('country') # reassign
```

```python
df.set_index('country', inplace=True)
```

```python
df
```

Select an individual element - return the capital of France?

```python
df['capital']['France']
```

```python
df.loc['France', 'capital']
```

```python
df['density'] = df['population'] / df['area'] * 10**6
df
```

```python
df['density'] > 300
```

```python
df[df['density'] > 300]
```

```python
df.sort_values('area')
```

What is the mean population for EU countries in this collection

```python
df.index
```

```python
eu_countries = df.index != 'United Kingdom'
eu_countries
```

```python
df.loc[eu_countries]
```

```python
df.loc[eu_countries]['population']
```

```python
df.loc[eu_countries]['population'].mean()
```

```python
df.loc[df.index != 'United Kingdom', 'population'].mean()
```
